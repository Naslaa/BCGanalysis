```python
import pandas as pd
```

```python
df= pd.read_excel('BCG.xlsx')
```

```python
df.head()
```

```python
import re
df.columns= [re.sub(r"\s*\(.*?\)", "", col) for col in df.columns]
```

```python
df.head()
```

```python
df['revenue growth %']=df.groupby(['Company'])['Total Revenue'].pct_change()*100
df['net income growth %']=df.groupby(['Company'])['Net Income'].pct_change()*100
```

<h2>Revenue Growth</h2>
Apple had a decrese in revenue by 1.98 % in 2024 compared to 2023, but there was a rise of 2.88% in revenue in 2023 compared to 2022. 
Microsoft had diped 6.43% in 2023 since 2022, and also extra 13.5% in 2024 from 2023. Revenue is decreasing more each year.
Tesle had 15.8% fall in the revenue in 2023, but it improved and had fall of only 0.9% in 2024 since 2023.
Overall all there is a decresing tread in all the companies.



<h2>Net Income Growth</h2>
Unlike revenue, Net income has increased by 2.8% in 2023 and by 3.47% in 2024 for Apple. But its not the same case for Microsoft, although there was a rise of 0.52% of net income in year 2023 compared to 2022, it dropped in 2024 with a sharp 15%. Its totally the opposite case for Tesla, although there was 15% drop in net income in 2023, Tesla turned it around with 109.3% total growth in net income in year 2024.
